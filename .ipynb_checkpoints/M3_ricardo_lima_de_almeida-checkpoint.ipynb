# Exercício 1

#### Criação

* Criar uma classe chamada Aluno que tenha os atributos: nome e nota
* Fazer o construtor da classe com dois parâmetros. Adicionar no construtor um print para informar que o aluno foi criado com sucesso.
* Adiciona o método string ( __str__ ), para que ao imprimir se veja a sua informação
* Criar um método chamado **qualificacao** que imprima se o aluno passou ou reprovou com base na sua nota

#### Experimentar

* Criar alguns objectos alunos
* Imprimir os objetos alunos criados (o objecto e seus atributos) e o método de qualificação

````python
# Fazer o exercício aqui
class Aluno:
    nota_max = 20
    nota_min = 0
    score = 0.7
    def __init__(self, nome, nota):
        self.nome = nome
        self.nota = nota
        print("O aluno foi criado com sucesso.")
        
    def __str__(self):
        return f"\nNome do aluno: {self.nome}\nNota: {self.nota}\n"
    
    def qualificacao(self):
        '''
        Este método verifica se o aluno foi aprovado segundo o score mínimo de aprovação.
        
        A nota máxima, nota mínima e score de aprovação(percentual) podem ser alterados
        conforme as regras da avaliação.
        
        Retorn uma string com o resultado "Aprovado" ou "Reprovado".
        '''
        
        if self.nota >= ((self.nota_max - self.nota_min) * self.score):
            return "Aprovado!\n"
        return "Reprovado.\n"

aluno_01 = Aluno("Pedro", 15)
aluno_02 = Aluno("Maria", 18)
aluno_03 = Aluno("Carlos", 12)
aluno_04 = Aluno("Joaquim", 14)

print(aluno_01, aluno_01.qualificacao(), sep="")
print(aluno_02, aluno_02.qualificacao(), sep="")
print(aluno_03, aluno_03.qualificacao(), sep="")
print(aluno_04, aluno_04.qualificacao(), sep="")
````

# Exercício 2

#### Criação

Sobre o exercício anterior fazer a seguinte modificação.

No exercício anterior, ao fazer um print as notas e nomes, vamos obter a nota e nome do aluno. Como pode ser considerado um problema de segurança, vamos encapsular o código para que não se possa aceder diretamente às variáveis e o acesso tenha de ser através de métodos.
* Fazer as variáveis de classe __privadas__. Lembrete: adicionar hífen duplo antes do nome da variável 
* Criar métodos __getter e setter__ poder aceder e modificar as variáveis de classe

#### Experimentar

* Criar alguns objectos alunos
* Comprovar que apenas podemos aceder às variáveis da classe através dos seus métodos get e set (Podemos aceder diretamente ao__nome e__nota? Podemos modifica-los?)

````python
# Realizaro exercício aqui
class Aluno:
    nota_max = 20
    nota_min = 0
    score = 0.7
    def __init__(self, nome, nota):
        self.__nome = nome
        self.__nota = nota
        print("O aluno foi criado com sucesso.")
        
    def __str__(self):
        return f"\nNome do aluno: {self.__nome}\nNota: {self.__nota}\n"
    
    def qualificacao(self):
        '''
        Este método verifica se o aluno foi aprovado segundo o score mínimo de aprovação.
        
        A nota máxima, nota mínima e score de aprovação(percentual) podem ser alterados
        conforme as regras da avaliação.
        
        Retorn uma string com o resultado "Aprovado" ou "Reprovado".
        '''
        
        if self.__nota >= ((self.nota_max - self.nota_min) * self.score):
            return "Aprovado!\n"
        return "Reprovado.\n"
    
    #Getters
    @property
    def nome(self):
        print("Chamado dentro do getter")
        return self.__nome
    
    @property
    def nota(self):
        print("Chamado dentro do getter")
        return self.__nota
    
    
    #Setters
    @nome.setter
    def nome(self, novo_nome):
        self.__nome = novo_nome
        
    @nota.setter
    def nota(self, nova_nota):
        self.__nota = nova_nota

aluno_01 = Aluno("Pedro", 15)
aluno_02 = Aluno("Maria", 18)
aluno_03 = Aluno("Carlos", 12)
aluno_04 = Aluno("Joaquim", 14)

print(aluno_01, aluno_01.qualificacao(), sep="")
print(aluno_02, aluno_02.qualificacao(), sep="")
print(aluno_03, aluno_03.qualificacao(), sep="")
print(aluno_04, aluno_04.qualificacao(), sep="")


print(aluno_02.nome) # Se chamar dentro do getter vai surgir um print alertando que foi dentro do getter.
#print(aluno_02.__nome) # Caso tente aceder diretamente apresenta o seguinte erro: AttributeError: 'Aluno' object has no attribute '__nome'.
#print(aluno_02.__nota) # O mesmo ocorre aqui se tentar aceder diretamente apresenta o erro.
print(aluno_04.nota) # É possível aceder através do getter novamente
aluno_02.nome = "Raquel" # É possível alterar através do setter
print(aluno_02.nome) # O nome foi alterado com sucesso.
aluno_02.__nota = 20 # Não resultou na alteração da nota quando tenta fazer diretamente.
print(aluno_02.nota) # A nota permaneceu a original
````

# Exercício 6

**Exercício sobre pontos, coordenadas e vetores no plano cartesiano. A seguir os conceitos básicos.**

### O plano cartesiano

Representa um espaço bidimensional (em 2 dimensões), formado por duas linhas perpendiculares, uma horizontal e outra vertical, que se intersetam num ponto. A linha horizontal é chamada eixo das abcissas ou __eixo X__, enquanto a vertical é chamada eixo das ordenadas ou simplesmente o __eixo Y__.
Quanto ao ponto de interseção, é conhecido como o ponto de origem O. __O plano está dividido em 4 quadrantes.__


<img src="./eixos.png"  />



### Pontos e coordenadas

O objetivo é descrever a posição dos **pontos** no plano sob a forma de **coordenadas**, que são formadas associando o valor do eixo X (horizontal) com o valor do eixo Y (vertical).


A representação de um ponto é simples: **P(X,Y)** onde X e Y são a distância horizontal (esquerda ou direita) e vertical (para cima ou para baixo) 
respetivamente, usando como referência o ponto de origem (0,0), mesmo no centro do plano.

<img src="./Puntos.png"  />


### Vetores no plano

Finalmente, um vetor no plano refere-se a um segmento orientado, gerado a partir de dois pontos diferentes.

Para efeitos práticos, é simplesmente uma linha formada a partir de um ponto inicial na direção de outro ponto final, pelo que se entende que um vetor tem comprimento e direção/sentido.

<img src="./Vector.png"  />


Nesta figura, podemos ver dois pontos A e B que poderíamos definir como se segue:
* **A(x1, y1)** => **A(2, 3)**
* **B(x2, y2)** => **B(7, 5)**

E o vetor seria representado como a diferença entre as coordenadas do segundo ponto em relação ao primeiro (o segundo menos o primeiro):
* **AB = (x2-x1, y2-y1)** => **(7-2, 5-3)** => **(5,2)** 

O que significa:

5 à direita e 2 para cima.

* Criar uma classe chamada **Retângulo** com dois pontos (início e fim) que irá formar a diagonal do retângulo.

*Pode facilmente identificar estes valores se tentar desenhar o quadrado a partir da sua diagonal. Se estiver perdido, tente desenhá-lo num pedaço de papel, vai vê-lo muito mais claramente! Lembre-se também que pode usar a função **abs()** para encontrar o valor absoluto de um número.*

#### Experiência
* Criar os pontos A(2, 3), B(5,5), C(-3, -1) e D(0,0) e imprimir. 
* Consultar a que quadrantes pertencem os pontos A, C e D.
* Consultar os vetores AB e BA.
* Consultar a distância entre os pontos 'A e B' e 'B e A'.
* Determinar qual dos 3 pontos A, B ou C está mais afastado da origem, ponto (0,0). 
* Criar um retângulo utilizando os pontos A e B.
* Consultar a base, altura e área do retângulo.

````python
# Fazer o exercício aqui
import math
class Ponto:
    def __init__(self, X = 0, Y = 0):
        self.X = X
        self.Y = Y
    
    def __str__(self):
        return f"({self.X},{self.Y})"
    
    def quadrante(self):
        if self.X == 0 and self.Y == 0:
            return "Origem"
        elif self.X > 0 and self.Y > 0:
            return "Quadrante I"
        elif self.X < 0 and self.Y > 0:
            return "Quadrante II"
        elif self.X < 0 and self.Y < 0:
            return "Quadrante III"
        else:
            return "Quadrante IV"
        
    def vetor(self,ponto2=None):
        if ponto2 == None:
            ponto2 = Ponto(0,0)
        return Ponto(ponto2.X - self.X, ponto2.Y - self.Y)
    
    def distancia(self, ponto2 = None):
        if ponto2 == None:
            ponto2 = Ponto(0,0)
        return math.sqrt((ponto2.X - self.X)**2 + (ponto2.Y - self.Y)**2)
    
class Retangulo:
    def __init__(self, p1 = Ponto(0, 0), p2 = Ponto(0, 0)):
        self.inicio = p1
        self.fim = p2
        print(f"\nRetângulo entre os pontos {self.inicio} e {self.fim} criado com sucesso.")
        
    def base(self):
        return abs(self.inicio.distancia(Ponto(self.fim.X, self.inicio.Y)))
    
    def altura(self):
        return abs(self.inicio.distancia(Ponto(self.inicio.X, self.fim.Y)))
    
    def area(self):
        return self.base() * self.altura()
    
def mais_afastado_origem(**kwargs):
    ponto = ""
    maior_distancia = -1
    for chave in kwargs:
        if kwargs[chave] > maior_distancia:
            maior_distancia = kwargs[chave]
            ponto = chave
    return f"{ponto}: {kwargs[ponto]:.2f}"
    
A, B, C, D = Ponto(2, 3), Ponto(5,5), Ponto(-3, -1), Ponto(0,0)
print(f"\nPonto A: {A}\nPonto B: {B}\nPonto C: {C}\nPonto D: {D}")
print(f"\nQuadrante do ponto A: {A.quadrante()}\nQuadrante do ponto B: {B.quadrante()}\nQuadrante do ponto C: {C.quadrante()}\nQuadrante do ponto D: {D.quadrante()}")
print(f"\nVetor AB: {A.vetor(B)}\nVetor BA: {B.vetor(A)}")
print(f"\nDistância AB: {A.distancia(B):.2f}\nDistância BA: {B.distancia(A):.2f}")

mais_afastado = mais_afastado_origem(A = A.distancia(), B = B.distancia(), C = C.distancia())
print(f"\nO ponto mais afastado da origem é o ponto {mais_afastado}")

retangulo1 = Retangulo(A, B)
print(f'''
Dados sobre o retangulo entre os pontos {retangulo1.inicio} e {retangulo1.fim}:

Base: {retangulo1.base()}
Altura: {retangulo1.altura()}
Área: {retangulo1.area()}

''')
````

````python
# Fazer o exercício aqui
class Veiculo:
    def __init__(self, cor, rodas):
        self.cor = cor
        self.rodas = rodas
    def __str__(self):
        return f"\nVeículo:\nCor: {self.cor}\nRodas: {self.rodas}"

class Carro(Veiculo):
    def __init__(self, cor, rodas, velocidade, cilindrada):
        super().__init__(cor, rodas)
        self.velocidade = velocidade
        self.cilindrada = cilindrada
    def __str__(self):
        return f"\nCarro:\nCor: {self.cor}\nRodas: {self.rodas}\nVelocidade: {self.velocidade} km/h\nCilindrada: {self.cilindrada} cc"
    

carro1 = Carro("Preto", 4, 180, 500)
print(carro1)
carro2 = Carro("Branco", 4, 210, 800)
print(carro2)
carro3 = Carro("Prata", 4, 240, 950)
print(carro3)
carro4 = Carro("Amarelo", 4, 300, 1500)
print(carro4)
````

O método deve percorrer a lista, procurar o nome do tipo da classe de cada objecto e encontrar os iguais ao argumento nome, e guardá-las numa nova lista. Esta nova lista deve ser devolvida ao programa principal onde será atravessada e o seu conteúdo exibido. A chamada no programa principal deve ter este aspeto: **``listaCarros = filtrar(lista, “Carro”)``**

````python
# Fazer o exercício aqui
class Veiculo:
    def __init__(self, cor, rodas):
        self.cor = cor
        self.rodas = rodas
    
    def __str__(self):
        atributos = (vars(self))
        string = type(self).__name__
        for atributo, info in atributos.items():
            string += f"\n{atributo.capitalize()}: {info}"
            if atributo.lower() == "velocidade":
                string += " km/h"
            if atributo.lower() == "cilindrada":
                string += " cc"
            if atributo.lower() == "carga":
                string += " ton"
        return string + "\n"
    
    def filtrar(lista_veiculos, veiculo):
        nova_lista = []
        for i in lista_veiculos:
            if type(i).__name__.lower() == veiculo.lower():
                nova_lista += [i]
        return nova_lista

class Carro(Veiculo):
    def __init__(self, cor, rodas, velocidade, cilindrada):
        super().__init__(cor, rodas)
        self.velocidade = velocidade
        self.cilindrada = cilindrada
    
    
class Bicicleta(Veiculo):
    def __init__(self, cor, rodas, tipo):
        super().__init__(cor, rodas)
        self.tipo = tipo
    
class Camiao(Carro):
    def __init__(self, cor, rodas, velocidade, cilindrada, carga):
        super().__init__(cor, rodas, velocidade, cilindrada)
        self.carga = carga
    

class Moto(Bicicleta):
    def __init__(self, cor, rodas, tipo, velocidade, cilindrada):
        super().__init__(cor, rodas, tipo)
        self.velocidade = velocidade
        self.cilindrada = cilindrada
    
    
#def catalogar(lista_veiculos):
#    for i in lista_veiculos:
#        print(i)


def catalogar(lista_veiculos, rodas = None):
    if rodas == None:
        for i in lista_veiculos:
            print(i)
        return
    count_veiculos = 0
    for j in lista_veiculos:
        if j.rodas == rodas:
            count_veiculos += 1
            print(j)
    if count_veiculos == 0:
        print(f"\nDesculpe, não foi encontrado veículos com {rodas} rodas.")
        return
    print(f"\nEncontrados {count_veiculos} veículos com {rodas} rodas:")
    

carro1 = Carro("Amarelo", 4, 300, 1500)
carro2 = Carro("Branco", 4, 240, 1000)
carro3 = Carro("Azul", 4, 180, 800)
bike1 = Bicicleta("Azul", 2, "desportivo")
bike2 = Bicicleta("Preto", 2, "urbano")
bike3 = Bicicleta("Vermelho", 2, "desportivo")
camiao1 = Camiao("Amarelo", 4, 300, 1500, 23)
camiao2 = Camiao("Cinza", 4, 170, 1500, 23)
camiao3 = Camiao("Branco", 4, 160, 900, 11)
moto1 = Moto("Azul", 2, "urbano", 160, 200)
moto2 = Moto("Verde", 2, "desportivo", 190, 220)
moto3 = Moto("Amarelo", 2, "desportivo", 210, 250)

lista_veiculos = []
lista_veiculos.extend([carro1, carro2, carro3, bike1, bike2, bike3, camiao1, camiao2, camiao3, moto1, moto2, moto3])
listaCarros = Carro.filtrar(lista_veiculos, "Carro")
print("\nFiltro de carros:\n")
catalogar(listaCarros)
listaCamiao = Camiao.filtrar(lista_veiculos, "Camiao")
print("\nFiltro de camiões:\n")
catalogar(listaCamiao)
listaBicicletas = Bicicleta.filtrar(lista_veiculos, "Bicicleta")
print("\nFiltro de bicicletas:\n")
catalogar(listaBicicletas)
listaMotos = Moto.filtrar(lista_veiculos, "Moto")
print("\nFiltro de motos:\n")
catalogar(listaMotos)
````
