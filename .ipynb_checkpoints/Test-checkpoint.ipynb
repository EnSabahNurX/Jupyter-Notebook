# Beautiful Soup example

```python
from requests import get
from bs4 import BeautifulSoup
```

```python
url = "https://www.imdb.com/title/tt0944947/episodes?season=1"
```

```python
response = get(url)
```

```python
print(response.text[:50])
```

```python
html_soup = BeautifulSoup(response.text, 'html.parser')
```

```python
episode_containers = html_soup.find_all('div',class_='info')
```

```python
print(episode_containers)
```

```python
print(episode_containers[0].a['title'])
print(episode_containers[1].a['title'])
print(episode_containers[2].a['title'])
```

```python
numbers = list(range(1,11))
print(numbers)
```

```python
for index,number in enumerate(numbers):
    numbers[index] **= 2
```

```python
print(numbers)
```
