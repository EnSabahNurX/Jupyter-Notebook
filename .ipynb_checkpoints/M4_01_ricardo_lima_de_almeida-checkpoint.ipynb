**1) Exercício de herança múltipla e polimorfismo**
* Criar a classe **Veiculos** com um construtor que inclua marca e modelo
* Esta classe **Veículos** também terá de incluir um método que se chame **reabastecer()** o qual imprime por ecrã “Este veículo tem de reabastecer gasolina”
* Criar a classe **VEletricos** com um construtor que inclua marca, modelo e autonomia
* Esta classe VEletricos também terá de incluir um método que se chame **reabastecer()** o qual imprime por ecrã “Este veículo tem de reabastecer eletricidade”
* Criar a classe **BicicletaEletrica** que herde de **Veiculos** e de **VEletricos**, mas dando prioridade a **VEletricos** (já que é um veículo mais elétrico que normal)
* Criar a classe **Quadriciclo** que herde de **Veiculos** e de **VEletricos**, mas dando prioridade a **Veículos** (já que apenas usa a eletricidade de modo pontual)
* Criar um objeto de **BicicletaEletrica** e outro de **Quadriciclo**
* Empregar as técnicas de polimorfismo aprendidas para conseguir que ao perguntar-lhe aos objetos que acabamos de criar por reabastecimento, ambos nos respondam adequadamente


```python
# Completar o exercício aqui
class Veiculos:
    def __init__(self, marca, modelo):
        self.marca = marca
        self.modelo = modelo
        
    def reabastecer(self):
        print("Este veículo tem de reabastecer gasolina")

class VEletricos:
    def __init__(self, marca, modelo, autonomia):
        self.marca = marca
        self.modelo = modelo
        self.autonomia = autonomia
        
    def reabastecer(self):
        print("Este veículo tem de reabastecer eletricidade")
        
class BicicletaEletrica(VEletricos, Veiculos):
    pass

class Quadriciclo(Veiculos, VEletricos):
    pass

bicicleta1 = BicicletaEletrica("Monark", "B43", 23)
quadriciclo1 = Quadriciclo("Honda", "FX45")

bicicleta1.reabastecer()
quadriciclo1.reabastecer()
print(vars(bicicleta1))
print(vars(quadriciclo1))
```

**2) Exercício de herança**
* Criar a superclasse **Pessoal_Universitario** com um único atributo que seja um dicionário que inclua (Keys): “id”, “nome” e “email” 
* Criar as classes **Escritorio**, **Professor** e **Aluno**. Todas herdaram de **Pessoal_Universitário** e os seus dados
* No caso de **Escritorio**, deve adicionar ao seu dicionário de dados o dado “Gabinete”
* No caso de **Professor**, deve adicionar ao seu dicionário de dados o dado “Especializacao”
* No caso de **Aluno**, deve adicionar ao seu dicionário de dados o dado “CreditosAprovados" (integer)  
* Criar métodos que sirvam para mostrar a informação de cada classe
* Criar objetos de todas as classes
* Mostrar a informação desses objetos e verificar o seu output.

```python
# Completar o exercício aqui
class Pessoal_Universitario:
    def __init__(self, **kwargs):
        self.dicionario = dict()
        for chave, valor in kwargs.items():  # percorre cada item do dicionário recebido em kwargs 
            self.dicionario[chave] = valor     # e adiciona o conjunto chave:valor no atributo dicionário da
            
    def mostrarInfo(self):
        string = type(self).__name__
        for atributo, info in self.dicionario.items():
            string += f"\n{atributo.capitalize()}: {info}"
        print(string + "\n")
        
class Escritorio(Pessoal_Universitario):
    def __init__(self, **kwargs):
        super().__init__(**kwargs)           # chama o init da classe pai com os pares de chave:valor recebidos
        for chave, valor in kwargs.items():  # em kwargs e depois herda este dicionário que é atributo da super classe
            if chave == "Gabinete":          # e percorre os pares de chave:valor restante e verifica se atende ao
                self.dicionario[chave] = valor # critério e então adiciona ao dicionário da classe
        
class Professor(Pessoal_Universitario):
    def __init__(self, **kwargs):
        super().__init__(**kwargs)
        for chave, valor in kwargs.items():
            if chave == "Especializacao":
                self.dicionario[chave] = valor
        
class Aluno(Pessoal_Universitario):
    def __init__(self, **kwargs):
        super().__init__(**kwargs)
        for chave, valor in kwargs.items():
            if chave == "CreditosAprovados":
                self.dicionario[chave] = valor
        
escritorio1 = Escritorio(id=11, nome="Central", email="central@universidade.pt", Gabinete="Gabinete de Apoio ao Estudante")
professor1 = Professor(id=54, nome="Nuno Pereira", email="nunopereira@universidade.pt", Especializacao="Engenharia de automação e controlo")
aluno1 = Aluno(id=619, nome="Miguel Oliveira", email="619aluno@universidade.pt", CreditosAprovados=350)

escritorio1.mostrarInfo()
professor1.mostrarInfo()
aluno1.mostrarInfo()
```

**3)  Escolher a temática que preferir e definir uma estrutura de classes e subclasses onde pratique a herança em pelo menos dois níveis (herança multinível). Criar uma estrutura de mínimo 5 classes, com os seus respetivos construtores e métodos. Criar alguns objetos e comprovar que a herança esteja a funcionar. Se puder, incluir a técnica de polimorfismo e herança múltipla.**

```python
# Completar o exercício aqui

# Classes para classificar jogos de modo simplificado
class Jogos:
    def __init__(self, nome, ano, categoria, valor, plataformas, interatividade, **kwargs):
        self.nome = nome
        self.ano = ano
        self.categoria = categoria
        self.valor = valor
        self.plataformas = plataformas
        self.interatividade = interatividade  # player vs player, cooperativo, equipa
        print(
            f"Jogo criado com sucesso. Classe de origem: {type(self).__name__}")

    def __str__(self):
        atributos = (vars(self))
        string = ""
        for atributo, info in atributos.items():
            if info:
                string += f"\n{atributo.capitalize()}: {info}"
        return string + "\n"


class Online(Jogos):
    def __init__(self, conexao, **kwargs):
        super().__init__(**kwargs) # "super" é usado para chamar o constructor da classe pai e envia como parâmetro todos os já recebidos de forma dinâmica usando os **kwargs
        self.conexao = conexao  # LAN, Internet

    def dica(self): # este método será chamado pelas classes netas e será dado prioridade conforme a classe herdada mais a esquerda
        print(f'''
        Vai começar o jogo {self.nome} do tipo {self.interatividade}
        no modo Online.
        (evite conexões lentas para ter uma melhor experiência durante a partida,
        sempre ajude os membros da equipa quando possível).
        
        ''')


class Offline(Jogos):
    def __init__(self, duracao, **kwargs):
        super().__init__(**kwargs) # novamente o "super" é usado para chamar o constructor da classe pai e envia como parâmetro todos os já recebidos de forma dinâmica usando os **kwargs
        self.duracao = duracao  # horas para concluir toda a história e narrativa

    def dica(self): # este método será chamado pelas classes netas e será dado prioridade conforme a classe herdada mais a esquerda
        print(f'''
        Vai começar o jogo {self.nome} do tipo {self.interatividade}
        no modo Offline.
        (não esqueça de guardar o progresso antes de sair do jogo).
        
        ''')


class Singleplayer(Offline, Online):
    def __init__(self, nome=None, ano=None, categoria=None, valor=None, plataformas=None, interatividade=None,
                 duracao=None, conexao=None): # atribui None por padrão e atualiza tudo que recebe como parâmetro
        super().__init__(nome=nome, ano=ano, categoria=categoria, valor=valor, plataformas=plataformas,
                         interatividade=interatividade, duracao=duracao,
                         conexao=conexao) # chama o constructor da classe acima e envia todos os parâmentros recebidos


class Multiplayer(Online, Offline):
    def __init__(self, nome=None, ano=None, categoria=None, valor=None, plataformas=None, interatividade=None,
                 duracao=None, conexao=None): # atribui None por padrão e atualiza tudo que recebe como parâmetro
        super().__init__(nome=nome, ano=ano, categoria=categoria, valor=valor, plataformas=plataformas,
                         interatividade=interatividade, duracao=duracao,
                         conexao=conexao) # chama o constructor da classe acima e envia todos os parâmentros recebidos


jogo1 = Singleplayer(nome="God of War", ano=2022, categoria="Ação, Aventura", valor="250 €",
                     plataformas="PC, Playstation", interatividade="Player vs Computer", duracao=85)
print(jogo1)
jogo1.dica() # aplicação de polimorfismo, método existe em 2 classes acima, será priorizada a primeira da esquerda
print(Singleplayer.__mro__) # verifica a ordem de prioridade em caso de existir métodos com o mesmo nome em mais de uma classe

jogo2 = Multiplayer(nome="League of Legends", ano=2022, categoria="MOBA, Estratégia", valor="0 €",
                    plataformas="PC, Android", interatividade="Equipa vs Equipa", conexao="Internet")
print(jogo2)
jogo2.dica() # aplicação de polimorfismo, método existe em 2 classes acima, será priorizada a primeira da esquerda
print(Multiplayer.__mro__) # verifica a ordem de prioridade em caso de existir métodos com o mesmo nome em mais de uma classe
```
